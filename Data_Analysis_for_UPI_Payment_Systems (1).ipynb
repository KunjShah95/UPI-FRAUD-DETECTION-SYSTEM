```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
from sklearn.base import BaseEstimator, ClassifierMixin
import numpy as np

X_train = np.array(X_train)
y_train = np.array(y_train)
# Custom wrapper for XGBClassifier
class CompatibleXGBClassifier(XGBClassifier, BaseEstimator, ClassifierMixin):
    def __sklearn_tags__(self):
        return {
            "non_deterministic": False,
            "requires_positive_y": False,
            "X_types": ["2darray"],
            "poor_score": False,
            "no_validation": False,
            "binary_only": False,
            "multioutput": False,
            "multioutput_only": False,
            "allow_nan": True,
            "stateless": False,
            "requires_fit": True,
        }

# Replace XGBClassifier with the compatible wrapper in models
if 'XGBClassifier' not in models:
    models['XGBClassifier'] = CompatibleXGBClassifier()

# Define hyperparameter grid
param_grid = {
    'DecisionTreeClassifier': {'max_depth': [10, 20, 50], 'min_samples_split': [2, 5, 10]},
    'RandomForestClassifier': {'n_estimators': [50, 100, 200], 'max_depth': [10, 20], 'min_samples_split': [2, 5]},
    'GradientBoostingClassifier': {'n_estimators': [50, 100], 'learning_rate': [0.05, 0.1], 'max_depth': [3, 5]},
    'XGBClassifier': {'n_estimators': [50, 100], 'learning_rate': [0.05, 0.1], 'max_depth': [3, 5]}
}

# Perform GridSearchCV
best_params = {}
for model_name, model in models.items():
    model_class_name = type(model).__name__

    if model_class_name in param_grid:
        grid_search = GridSearchCV(
            estimator=model,
            param_grid=param_grid[model_class_name],
            cv=5,
            scoring='f1',
            verbose=2,
            error_score='raise'
        )
        grid_search.fit(X_train, y_train)
        best_params[model_name] = grid_search.best_params_
    else:
        print(f"Warning: No hyperparameters found for {model_name} in param_grid.")

# Print best parameters
for model_name, params in best_params.items():
    print(f"Best hyperparameters for {model_name}: {params}")
```
